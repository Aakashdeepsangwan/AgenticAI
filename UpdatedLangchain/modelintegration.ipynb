```python
from langchain_anthropic import ChatAnthropic
import os
from dotenv import load_dotenv


load_dotenv()
api_key = os.getenv("claudeAPI")
```

```python
llm = ChatAnthropic(
    model  = "claude-haiku-4-5-20251001",
    temperature = 0.7,
    api_key= api_key
)
```

```python
result = llm.invoke("Tell me in 200 words, How to ace an Network Automation intership interview")
```

```python
for chunk in llm.stream("Tell me in 200 words, what's AgenticAI") :
    print(chunk.text, end = "|", flush = True)
```

Batch is a collection of independent requests to a model can significantly improve the performance and reduce cost, as the processing can be done in parallel

```python
responses = llm.batch(
    [
        "Who is the best footballer in the world right now?",
        "What does an AI/ML engineer?",
        "Who is the best coder in the world right now"
    ]
)

for response in responses :
    print(response)
```
